```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df2=df.groupby('day').sum(numeric_only=True) # sum per day
df2.drop('size',inplace=True,axis=1) # sum of size column is not relevant
df2['percent'] = df2['tip']/df2['total_bill']*100 # add percents
df2
```

```python
df3=df.groupby('smoker').sum(numeric_only=True)
df3['percent'] = df3['tip']/df3['total_bill']*100
df3
```

```python
days=pd.get_dummies(df['day']).astype(int)
days.sample(5)
```

```python
days=pd.get_dummies(df['day']).astype(int)
df = pd.concat([df,days],axis=1)
```

```python
times=pd.get_dummies(df['time']).astype(int)
df = pd.concat([df,times],axis=1)
```

```python
df
```

```python
X = df[['total_bill','size']]
Y = df[['tip']]
```

```python
X
```

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
X_train, X_test , y_train , y_test = train_test_split(X,Y,test_size=0.25,random_state=26)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
predictions=regressor.predict(X_test)
```
